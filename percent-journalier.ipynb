```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('HistoricalData_1698759211717.csv', index_col='Date')
#df['Date'] = df['Date'].to_datetime

for c in ['Close/Last', 'Open', 'High', 'Low'] :
#    df[c] = df[c].str[1:].astype(float)
    df[c] = df[c].str.replace('$', '').astype(float)

df['Open'] = (df['Open'].pct_change()*100)

df['Open'].head(100).plot()
df['Open']=df['Open'].head(100)
```

```python
plt.plot(df['Open'], color='blue', label='Valeur')

positive_values = df[df['Open'] >= 0]['Open']
negative_values = df[df['Open'] < 0]['Open']

if not positive_values.empty:
    plt.plot(positive_values.index, positive_values, 'go', label='Positive')

if not negative_values.empty:
    plt.plot(negative_values.index, negative_values, 'ro', label='Ngative')

plt.xlabel('Date')
plt.ylabel('Pourcentage daugmentation')

plt.legend()

plt.show()
```
